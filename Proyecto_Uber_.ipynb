### Creamos una columna aparte

```python
base=np.zeros((14,24))
base
```

```python
fila = np.arange(0,14)
```

```python
fila
```

```python
fila=np.expand_dims(fila,axis=0)
fila
```

```python
fila=fila.T
```

```python
dia= base + fila
dia
```

```python
dia=dia.reshape((1,336))
dia
```

```python
dia = np.squeeze(dia)
dia
```

```python
dia.dtype
```

```python
dia=dia.astype(int)
```

### aplicamos la columna al DataFrame

```python
df['dia_formal'] = dia
```

```python
type(df['dia_formal'])
```

```python
df
```

```python
pregunta1=df.groupby('dia_formal')['Completed Trips '].sum().to_frame().reset_index()
pregunta1
```

### RESP 1

```python
pregunta1[pregunta1['Completed Trips '] == pregunta1['Completed Trips '].max()]
```

```python
plt.style.use("seaborn-dark-palette")
```

```python
plt.bar(pregunta1['dia_formal'],pregunta1['Completed Trips '], facecolor = 'red')
plt.xlabel('dia_formal')
plt.ylabel('Completed Trips')
plt.grid(True)

plt.show()
```
